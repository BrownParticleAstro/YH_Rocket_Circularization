```python
from radial_rocket import RadialBalance, DiscretiseAction
import rocket_gym
from DQN import DeepQNetwork

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
model = DeepQNetwork(dims=[2, 128, 128, 3],
                        epsilon=1.0, epsilon_decay=.1, gamma=.95,
                        memory=100000, start_updating=10000,
                        batch_size=32, learning_rate=1e-3, descent_frequency=16, update_frequency=1,
                        use_target=True, target_frequency=8)

env = DiscretiseAction(RadialBalance())
model.train(env, episodes=1000, render_frequency=100)
model.save('./radial_2/')

# model.load('./dqn_test_1/')

model.simulate(env, render=True, evaluation=True)
```

```python
with rocket_gym.make('RocketCircularization-v0') as env:
    env = rocket_gym.RadialObservation(
        rocket_gym.RadialThrust(
            rocket_gym.PolarizeAction(env)))
    model.simulate(env, render=True, evaluation=True)
```

```python
rs = np.linspace(0.5, 1.5, 100)
rdts = np.linspace(-10, 10, 100)
rm, rdtm = np.meshgrid(rs, rdts)
inputs = np.hstack(
    (rm.reshape(-1, 1), rdtm.reshape(-1, 1)))
action = tf.reshape(tf.argmax(model.q_net(inputs), axis=-1), (100, 100))
max_Q = tf.reshape(tf.reduce_max(model.q_net(inputs), axis=-1), (100, 100))

cs = plt.contourf(rm, rdtm, max_Q, label='max Q value')
plt.colorbar(cs)
plt.xlabel('radius')
plt.ylabel('r dot')
plt.legend()
plt.show()
```
